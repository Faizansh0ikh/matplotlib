Matplotlib.

Matplotlib is a popular data visualization library in Python. 
It provides a wide range of tools for creating various types of plots and graphs to visualize data in a clear and informative way. 
Matplotlib is widely used in fields such as data analysis, scientific research, engineering, and more

Line Plot: A basic plot where data points are connected by straight lines.
Bar Plot: Displays data using rectangular bars, often used for comparing categorical data.
Scatter Plot: Represents individual data points as dots, useful for showing the relationship between two continuous variables.
Histogram: Shows the distribution of a dataset by grouping data into bins and plotting the frequency of each bin.
Pie Chart: Displays data as sectors of a circle, useful for showing parts of a whole or proportions.

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. 
This is particularly useful when you want to display multiple plots side by side in a single figure,
making it easier to compare different visualizations.

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y, marker = "o", c = "y")
plt.show()
```

```python

# A box plot is a statistical visualization that provides a compact way to represent the distribution of a dataset. It displays the five-number summary of the data, which includes the minimum, first quartile (25th percentile), median (second quartile), third quartile (75th percentile), and maximum. The box plot consists of a rectangular box, a line (whisker) extending from the box, and possibly individual data points or outliers.

# Box plots are used to:

# Identify the central tendency (median) and spread (interquartile range) of the data.
# Detect outliers and extreme values in the dataset.
# Compare distributions between different categories or groups.
# Visualize the symmetry and skewness of the data.






box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1, box2], labels= ['Box 1', 'Box 2'])
plt.show()
```
